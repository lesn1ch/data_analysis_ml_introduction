### Кластеризация

```python
import seaborn as sns
import pandas as pd
from sklearn.cluster import KMeans

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/students.csv')
df.info()
```

```python
df_cut = df[['Weight', 'Shoe size']]
df_cut = df_cut.dropna()
```

```python
kmeans = KMeans(n_clusters=2, random_state=1)
kmeans.fit(df_cut)
df_cut['Cluster'] = kmeans.labels_
sns.scatterplot(data=df_cut, x='Shoe size', y='Weight', hue='Cluster')
```
